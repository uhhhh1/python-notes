# Python Programming: A General Overview

## Introduction
<br>
Python is a high-level, interpreted programming language known for its readability and simplicity. It's widely used in various fields like web development, data science, artificial intelligence, and more.

### Variables
variables are used to store data values. In python, you don't need to declare the type of a variable explicity.
```
# Example of variable assignment

x = 400    # Interger
y = 4.13   # Float
name = "sigma" # string
```

### Data Types
Common data types in Python include:

* **intergers** (int): whole numbers, e.g., 3, 100
*  **floating-point numbers** (float):Decimal numbers, e.g., 1.23, 9.65
*  **Strings** (str): Text data, e.g., "Hello, World!"
*  **Booleans**(bool): Logical values, True or false
*  **list** (list): Ordered, mutable collections, e.g., (4, 5, 6)
*  **sets**(set): unordered collections of unique items, e.g., {5, 3, 7}

## Control Flow
### If Statements

used to execute code based on a condition
# Example of an if statement 
```
age = 21
if age >= 21:
    print("You are an adult.")
else:
    print("You are a minor.")
```

### loops

loops are used to repeat a block of code multiple times.

#### for loops

Used for iterating over a sequence (like a list or a range).

# For Loop example

```
for i in range(10):
    print("Iteration", i)
```

#### While Loops

Repeats as long as condition is true.

```
# While loop example
count = 0
while count < 5:
    print("Count is," count)
count += 1
```

## Lists

Lists are ordered, mutable collections of items.

```
# creating a list
fruits = ["apple", "cherry", "watermelon"]

#Accessing list items
print(fruits[0])  # Outputs: apple

# Modifying list items
fruits[1] = "blueberry"

#adding items to a list
fruits.append("date")

# Removing items from a list
fruits.remove ("apple")
```

### Common List Methods

* append(item) :Adds an item to the end of a list
* insert (index, item) :inserts an item at a specified index.
* remove(item) :Removes the first occurence of an item.
* pop(index) :Removes and returns the item at the given index.
* sort() :Sorts the list
* reverse() :Reverses the list

## Dictionaries

dictionaries are used to store data values in key-value pairs.

```
# Creating a dictionary
person = { "name": "Sigma", "age": 21 }

# Accessing dictionary items
print(person["name"]) # Outputs: Sigma

#Modifying dictionary items
person["age"] = 21

# Adding a new key-value pair
person["city"] = "New York"
```

### Common Dictionary Methods

* keys() :Returns a list of keys.
* values() :Returns a list of values
* items() :Returns a list of key-value pairs.
* get(key) :Returns the value for the specified key.
